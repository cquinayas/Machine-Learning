# Ejercicio Red Neuronal funci√≥n XOR

```python
import tensorflow
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
import matplotlib.pyplot as plt
```

```python
# cargamos las 4 combinaciones de las compuertas XOR
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# y estos son los resultados que se obtienen, en el mismo orden
target_data = np.array([[0],[1],[1],[0]], "float32")

# datos de test
test_data = np.array([[0,0.01],[0.02,0.95],[0.85,0.01],[0.9,1]], "float32")

```

```python
model = Sequential()
model.add(Dense(3, input_dim=2, activation='relu'))#tanh
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#model.compile(loss='mean_squared_error',#binary_crossentropy 
              #optimizer='adam',#SGD 
              #metrics=['binary_accuracy']) 
```

```python
fit_history=model.fit(training_data, target_data, epochs=1000)
```

```python
# evaluamos el modelo
scores = model.evaluate(test_data, target_data)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
print (model.predict(test_data).round())
```

Vemos que las predicciones se corresponden con las salidas deseadas de la funci√≥n XOR.
